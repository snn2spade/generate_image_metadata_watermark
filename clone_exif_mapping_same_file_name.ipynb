```python
import os
from pathlib import Path
import piexif
```

```python
# original exif folder
in_folder_path_1 = "/Users/snn2spade/Desktop/Pai Trip 29-2 Feb 19/DCIM2"
# target replace exif folder
in_folder_path_2 = "/Users/snn2spade/Desktop/Pai Trip 29-2 Feb 19/postcard copy"

file_list_1 = [] 
for file in os.listdir(in_folder_path_1):
    file_list_1.append(Path(in_folder_path_1 +"/"+file))
    
file_list_2 = [] 
for file in os.listdir(in_folder_path_2):
    file_list_2.append(Path(in_folder_path_2 +"/"+file))
    
for file_2 in file_list_2:
    for file_1 in file_list_1:
        if file_1.stem == file_2.stem:
            piexif.transplant(str(file_1), str(file_2))
            print("Transplant completed [{}]".format(file_1.stem))
```
